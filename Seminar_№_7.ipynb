 _**3-я гипотеза**_ из задания № 2 набрала больше всего баллов (**392**):<br>
Если предусмотреть скидки  за последующие заказы, то число посещений и процент клиентов,<br>
которые возвращаются для совершения повторных покупок увеличатся.<br>

<center><big><big> План эксперимента.</big></big></center>

 **1.  Цель планируемого А/В-теста:** повысить процент клиентов, которые возвращаются для совершения<br> 
 повторных покупок за счет посещений пользователей со скидкой по промокоду.
 *** 
 **2.  Метрика:** будем оценивать **CRR**.
 ***
 **3. Гипотеза:** Ожидаем, что получение пользователем промокода на скидку  вызовет рост **CRR** на **10%** и более.
 ***
 **4. Настройка эксперимента:** 
 + **параметры для А/В-групп:** тестируем на всех пользователях, <br>
 контрольная - без изменений, тестовая - получает промокод на скидку при следующей покупке.
 + **проверка выборок на однородность:** запускаем предварительно  А/А-тест для оценки целевых метрик<br>
 на отсутствие статистически значимых различий.
 + **Рассчёт объема выборок и длительности проведения эксперимента:** объём выборок рссчитаем<br> калькулятором по ссылке: https://mindbox.ru , длительность проведения - отношение объёма выборок<br>
 и среднего количества пользователей в день.
 ***
 **5. Проведение эксперимента:** главное правило — не делать преждевременных выводов ,<br>
 чтобы избежать Peeking Problem (проблему «подглядывания») , надо предусмотреть невозможность<br>
 получать результаты теста до его окончания всем кроме аналитика.
  ***
 **6. Анализ результатов:** для сравнения результатов воспользуемся калькулятором по ссылке:<br> https://www.evanmiller.org/ab-testing.<br>
 
**В случае успешности - применяем изменения для всех пользователей,<br>
при неудаче - оставляем прежнюю версию.**
